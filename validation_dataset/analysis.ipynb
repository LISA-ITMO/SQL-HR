```python
import re
import pandas as pd

# ==========
# 1) INPUT / OUTPUT PATHS
# ==========
INPUT_PATH = "raw/dataset_kmu.csv"
VACANCIES_OUT = "prepared/vacancies.csv"
QUERIES_OUT = "prepared/queries.csv"

# ==========
# 2) HELPERS
# ==========
def clean_text(x):
    """Удаляет html-подобные теги, лишние пробелы и переводит NaN в пустую строку."""
    if pd.isna(x):
        return ""
    x = str(x)
    x = re.sub(r"<[^>]+>", " ", x)       # убираем теги вроде <highlighttext>
    x = x.replace("\xa0", " ")
    x = re.sub(r"\s+", " ", x).strip()
    return x


# ==========
# 3) READ SOURCE DATA
# ==========
df = pd.read_csv(INPUT_PATH)

# Если есть служебная колонка индекса — удаляем
unnamed_cols = [c for c in df.columns if c.lower().startswith("unnamed")]
if unnamed_cols:
    df = df.drop(columns=unnamed_cols)

# ==========
# 4) CLEAN ALL TEXT COLUMNS
# ==========
for col in df.columns:
    df[col] = df[col].apply(clean_text)

# ==========
# 5) OPTIONAL: MANUAL DOMAIN FIXES
# ==========
# Здесь можно руками поправить явно ошибочные домены по названию вакансии.
# При необходимости просто добавляй новые строки в словарь.
DOMAIN_FIXES = {
    "QA инженер": "qa",
    # "Системный аналитик": "analyst",
    # "Старший системный аналитик": "analyst",
    # "Бизнес-аналитик": "analyst",
    # "Младший аналитик": "analyst",
}

if "должность" in df.columns and "domen" in df.columns:
    df["domen"] = df.apply(
        lambda row: DOMAIN_FIXES.get(row["должность"], row["domen"]),
        axis=1
    )

# ==========
# 6) BUILD VACANCIES TABLE
# ==========
# Переименовываем колонки в более удобный формат
rename_map = {
    "должность": "job_title",
    "город": "city",
    "зарплата": "salary",
    "валюта зарплаты": "salary_currency",
    "зарплата за период времени": "salary_period",
    "занятость": "employment",
    "формат работы": "work_format",
    "часы работы": "work_hours",
    "график по дням": "work_schedule",
    "опыт работы": "experience",
    "требования": "requirements",
    "обязанности": "responsibilities",
    "domen": "domain",
}

vacancies = df.rename(columns=rename_map).copy()

# Добавляем стабильный vacancy_id
vacancies.insert(0, "vacancy_id", [f"VAC_{i:04d}" for i in range(1, len(vacancies) + 1)])

# Оставляем нужные колонки в понятном порядке
vacancy_cols = [
    "vacancy_id",
    "job_title",
    "domain",
    "city",
    "salary",
    "salary_currency",
    "salary_period",
    "employment",
    "work_format",
    "work_hours",
    "work_schedule",
    "experience",
    "requirements",
    "responsibilities",
]

# Оставляем только существующие колонки
vacancy_cols = [c for c in vacancy_cols if c in vacancies.columns]
vacancies = vacancies[vacancy_cols].copy()

# Дополнительно можно убрать совсем пустые строки, если вдруг есть
vacancies = vacancies.drop_duplicates().reset_index(drop=True)
```

```python
vacancies.columns
```

```python
# ==========
# 7) BUILD QUERIES TABLE
# ==========
query_type_cols = ["strict", "normal", "fuzzy", "complex"]
existing_query_cols = [c for c in query_type_cols if c in df.columns]

queries = (
    vacancies[["vacancy_id", "job_title", "domain"]]
    .join(df[existing_query_cols])
    .melt(
        id_vars=["vacancy_id", "job_title", "domain"],
        value_vars=existing_query_cols,
        var_name="query_type",
        value_name="query_text"
    )
)

# Чистим пустые query
queries["query_text"] = queries["query_text"].apply(clean_text)
queries = queries[queries["query_text"] != ""].copy()

# Добавляем query_id
queries.insert(0, "query_id", [f"Q_{i:05d}" for i in range(1, len(queries) + 1)])

# Упорядочим колонки
query_cols = [
    "query_id",
    "vacancy_id",
    "job_title",
    "domain",
    "query_type",
    "query_text",
]
queries = queries[query_cols].reset_index(drop=True)

```

```python
# ==========
# 8) SAVE
# ==========
vacancies.to_csv(VACANCIES_OUT, index=False, encoding="utf-8-sig")
queries.to_csv(QUERIES_OUT, index=False, encoding="utf-8-sig")

# ==========
# 9) QUICK CHECK
# ==========
print("Vacancies shape:", vacancies.shape)
print("Queries shape:", queries.shape)

print("\nVacancies preview:")
print(vacancies.head(3))

print("\nQueries preview:")
print(queries.head(8))

print("\nQuery type distribution:")
print(queries["query_type"].value_counts(dropna=False))

print("\nDomain distribution in vacancies:")
print(vacancies["domain"].value_counts(dropna=False))
```
