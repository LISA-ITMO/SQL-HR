```python
import re
import pandas as pd

# ==========
# 1) INPUT / OUTPUT PATHS
# ==========
INPUT_PATH = "raw/dataset_kmu.csv"
VACANCIES_OUT = "prepared/vacancies.csv"
QUERIES_OUT = "prepared/queries.csv"

# ==========
# 2) HELPERS
# ==========
def clean_text(x):
    """Удаляет html-подобные теги, лишние пробелы и переводит NaN в пустую строку."""
    if pd.isna(x):
        return ""
    x = str(x)
    x = re.sub(r"<[^>]+>", " ", x)       # убираем теги вроде <highlighttext>
    x = x.replace("\xa0", " ")
    x = re.sub(r"\s+", " ", x).strip()
    return x


# ==========
# 3) READ SOURCE DATA
# ==========
df = pd.read_csv(INPUT_PATH)

# Если есть служебная колонка индекса — удаляем
unnamed_cols = [c for c in df.columns if c.lower().startswith("unnamed")]
if unnamed_cols:
    df = df.drop(columns=unnamed_cols)

# ==========
# 4) CLEAN ALL TEXT COLUMNS
# ==========
for col in df.columns:
    df[col] = df[col].apply(clean_text)

# ==========
# 5) OPTIONAL: MANUAL DOMAIN FIXES
# ==========
# Здесь можно руками поправить явно ошибочные домены по названию вакансии.
# При необходимости просто добавляй новые строки в словарь.
DOMAIN_FIXES = {
    "QA инженер": "qa",
    # "Системный аналитик": "analyst",
    # "Старший системный аналитик": "analyst",
    # "Бизнес-аналитик": "analyst",
    # "Младший аналитик": "analyst",
}

if "должность" in df.columns and "domen" in df.columns:
    df["domen"] = df.apply(
        lambda row: DOMAIN_FIXES.get(row["должность"], row["domen"]),
        axis=1
    )

# ==========
# 6) BUILD VACANCIES TABLE
# ==========
# Переименовываем колонки в более удобный формат
rename_map = {
    "должность": "job_title",
    "город": "city",
    "зарплата": "salary",
    "валюта зарплаты": "salary_currency",
    "зарплата за период времени": "salary_period",
    "занятость": "employment",
    "формат работы": "work_format",
    "часы работы": "work_hours",
    "график по дням": "work_schedule",
    "опыт работы": "experience",
    "требования": "requirements",
    "обязанности": "responsibilities",
    "domen": "domain",
}

vacancies = df.rename(columns=rename_map).copy()

# Добавляем стабильный vacancy_id
vacancies.insert(0, "vacancy_id", [f"VAC_{i:04d}" for i in range(1, len(vacancies) + 1)])

# Оставляем нужные колонки в понятном порядке
vacancy_cols = [
    "vacancy_id",
    "job_title",
    "domain",
    "city",
    "salary",
    "salary_currency",
    "salary_period",
    "employment",
    "work_format",
    "work_hours",
    "work_schedule",
    "experience",
    "requirements",
    "responsibilities",
]

# Оставляем только существующие колонки
vacancy_cols = [c for c in vacancy_cols if c in vacancies.columns]
vacancies = vacancies[vacancy_cols].copy()

# Дополнительно можно убрать совсем пустые строки, если вдруг есть
vacancies = vacancies.drop_duplicates().reset_index(drop=True)
```

```python
vacancies.columns
```

```python
# ==========
# 7) BUILD QUERIES TABLE
# ==========
query_type_cols = ["strict", "normal", "fuzzy", "complex"]
existing_query_cols = [c for c in query_type_cols if c in df.columns]

queries = (
    vacancies[["vacancy_id", "job_title", "domain"]]
    .join(df[existing_query_cols])
    .melt(
        id_vars=["vacancy_id", "job_title", "domain"],
        value_vars=existing_query_cols,
        var_name="query_type",
        value_name="query_text"
    )
)

# Чистим пустые query
queries["query_text"] = queries["query_text"].apply(clean_text)
queries = queries[queries["query_text"] != ""].copy()

# Добавляем query_id
queries.insert(0, "query_id", [f"Q_{i:05d}" for i in range(1, len(queries) + 1)])

# Упорядочим колонки
query_cols = [
    "query_id",
    "vacancy_id",
    "job_title",
    "domain",
    "query_type",
    "query_text",
]
queries = queries[query_cols].reset_index(drop=True)

```

```python
vacancies.columns
```

```python
# ==========
# 8) SAVE
# ==========
vacancies.to_csv(VACANCIES_OUT, index=False, encoding="utf-8-sig")
queries.to_csv(QUERIES_OUT, index=False, encoding="utf-8-sig")

# ==========
# 9) QUICK CHECK
# ==========
print("Vacancies shape:", vacancies.shape)
print("Queries shape:", queries.shape)

print("\nVacancies preview:")
print(vacancies.head(3))

print("\nQueries preview:")
print(queries.head(8))

print("\nQuery type distribution:")
print(queries["query_type"].value_counts(dropna=False))

print("\nDomain distribution in vacancies:")
print(vacancies["domain"].value_counts(dropna=False))
```

```python
candidates['ready_to_relocate'] = candidates['ready_to_relocate'].apply(lambda x: 'не готов к переезду' if not x else 'готов к переезду')
candidates['ready_for_business_trips'] = candidates['ready_for_business_trips'].apply(lambda x: 'не готов к командировкам' if not x else 'готов к командировкам')
candidates['has_car'] = candidates['has_car'].apply(lambda x: 'нет машины' if not x else 'есть машина')
candidates = candidates.drop('resume_updated_at', axis=1)


def safe_bool_text(x, true_text, false_text, na_text='не указано'):
    if pd.isna(x):
        return na_text
    return true_text if bool(x) else false_text

def create_full_profile(row):
    parts = [
        f"Желаемая позиция: {row.get('desired_position', 'не указано')}",
        f"Последняя должность: {row.get('last_job_title', 'не указано')}",
        f"Последняя компания: {row.get('last_company', 'не указано')}",
        f"Город: {row.get('city', 'не указано')}",
        f"Тип занятости: {row.get('employment_type', 'не указано')}",
        f"График работы: {row.get('work_schedule', 'не указано')}",
        f"Ожидаемая зарплата: {row.get('expected_salary_rub', 'не указано')} руб.",
        f"Готовность к переезду: {safe_bool_text(row.get('ready_to_relocate'), 'готов к переезду', 'не готов к переезду')}",
        f"Готовность к командировкам: {safe_bool_text(row.get('ready_for_business_trips'), 'готов к командировкам', 'не готов к командировкам')}",
        f"Образование: {row.get('education_level_and_university', 'не указано')}",
        f"Опыт работы: {row.get('work_experience', 'не указано')}",
    ]
    return " ".join(map(str, parts))

candidates['full_resume'] = candidates.apply(create_full_profile, axis=1)
```

```python
quieries = pd.read_csv('prepared/queries.csv')
```

```python
# 1. Считаем длину текста в символах
lengths = quieries['query_text'].str.len()

# 2. Считаем квартили
q1 = lengths.quantile(0.25) # Конец "коротких"
q3 = lengths.quantile(0.75) # Начало "длинных"
iqr = q3 - q1               # Сам интерквартильный размах

print(f"Граница коротких (Q1): {q1} симв.")
print(f"Медиана (Q2): {lengths.median()} симв.")
print(f"Граница длинных (Q3): {q3} симв.")
print(f"IQR: {iqr}")
```

```python
quieries['query_length_bucket'] = quieries['query_text'].apply(lambda x: 'short' if len(x) < 100 else 'medium' if 100 <= len(x) <= 170 else 'long')
```

```python
def get_balanced_gold_subset(df, target_counts):
    df = df.copy()
    gold_indices = []

    for q_type, total_needed in target_counts.items():
        # Все запросы нужного типа (например, 'strict')
        type_subset = df[df['query_type'] == q_type]

        # 1. Пытаемся взять равномерно по длинам
        lengths = ['short', 'medium', 'long']
        per_length = total_needed // 3

        type_selected_indices = []

        for length in lengths:
            pool = type_subset[type_subset['query_length_bucket'] == length]
            # Берем сколько нужно, но не больше, чем есть в наличии
            n = min(len(pool), per_length)
            if n > 0:
                selected = pool.sample(n=n, random_state=42)
                type_selected_indices.extend(selected.index.tolist())

        # 2. Если после этого набрали меньше, чем total_needed (например, не хватило 'long')
        # Добираем остаток из всех оставшихся запросов этого ТИПА (любой длины)
        remaining_needed = total_needed - len(type_selected_indices)
        if remaining_needed > 0:
            # Исключаем тех, кого уже выбрали
            remaining_pool = type_subset.drop(type_selected_indices)
            n_extra = min(len(remaining_pool), remaining_needed)
            if n_extra > 0:
                extra_selected = remaining_pool.sample(n=n_extra, random_state=42)
                type_selected_indices.extend(extra_selected.index.tolist())

        gold_indices.extend(type_selected_indices)

    # Ставим флаг
    df['is_gold'] = False
    # Используем list(set()), чтобы точно не было дублей в индексах
    df.loc[list(set(gold_indices)), 'is_gold'] = True

    return df

# Твои цели из ТЗ
targets = {
    'strict': 8,
    'normal': 16,
    'fuzzy': 8,
    'complex': 8
}

gold_dataset = get_balanced_gold_subset(quieries, targets)
```
