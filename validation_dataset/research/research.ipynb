```python
candidates = pd.read_csv('../prepared/pooled_candidates.csv')
queries = pd.read_csv('../prepared/queries.csv')
```

```python
queries = queries[['query_id', 'job_title', 'query_text']]
```

```python
data_check = (
    vectors_gpt_5_mini_and_gpt_5
    .merge(queries, on='query_id', how='left')
    .merge(candidates, on='candidate_id', how='left')
    )

data_check = data_check[data_check['relevance_score_gpt_5'] == 2]
data_check.reset_index(drop=True).to_excel('data/vectors_2.xlsx', index=False)
```
