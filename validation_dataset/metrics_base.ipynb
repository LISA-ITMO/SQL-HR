# Metrics

```python
# concatenated
retrieval_results_dublicates = pd.read_csv('all_search_systems_concat/retrieval_results_dublicates.csv')
# 'золотые' запросы, которые будем размечать
gold_subset = pd.read_csv('llm_judge/gold_subset.csv')
# это предыдущий файл + доделали те кейсы, которы не отработали (около 30 строчек)
judgments = pd.read_csv('llm_judge/judgments.csv')
```

```python
gold_query_ids = gold_subset.loc[gold_subset["is_gold"] == True, "query_id"].unique()
gold_query_ids
```

```python
retrieval_results_dublicates_selected = retrieval_results_dublicates[retrieval_results_dublicates['query_id'].isin(gold_query_ids)]
retrieval_results_dublicates_selected.shape
```

```python
retrieval_results_dublicates_selected.columns
```

```python
judgments.columns
```

```python
sql_bm_vector_metrics = retrieval_results_dublicates_selected.merge(
    judgments[['query_id', 'candidate_id', 'relevance_score', 'judge_model', 'judge_prompt', 'judge_comment']],
    on=["query_id", "candidate_id"],
    how="left",
)
sql_bm_vector_metrics.shape
```

```python
sql_bm_vector_metrics.columns
```

```python
missing_labels = sql_bm_vector_metrics["relevance_score"].isna().sum()
assert missing_labels == 0, (
    f"После merge осталось {missing_labels} строк без relevance_score. "
    f"Проверь соответствие query_id/candidate_id между retrieval и judgments."
)

# relevance_score должен быть только 0/1/2
bad_labels = sorted(set(sql_bm_vector_metrics["relevance_score"].dropna().unique()) - {0, 1, 2})
assert not bad_labels, f"Обнаружены неожиданные relevance_score: {bad_labels}"

# Сортировка для удобства
sql_bm_vector_metrics = sql_bm_vector_metrics.sort_values(
    by=["query_id", "search_system", "rank", "candidate_id"]
).reset_index(drop=True)
```

```python
sql_bm_vector_metrics.to_csv('final/sql_bm_vector_metrics.csv', index=False)
```

```python
distribution = (
    sql_bm_vector_metrics.groupby(["search_system", "relevance_score"])
    .size()
    .unstack(fill_value=0)
    .reindex(columns=[0, 1, 2], fill_value=0)
    .astype(int)
)
distribution
```

# create pooled candidates for next systems

```python
from pathlib import Path
import pandas as pd

# =========================
# Пути
# =========================
BASE = Path(".")

candidates_path = BASE / "raw" / "candidates.csv"
retrieval_nodup_path = BASE / "all_search_systems_concat" / "retrieval_results_no_dublicates.csv"
gold_subset_path = BASE / "llm_judge" / "gold_subset.csv"

output_pooled_candidates_path = BASE / 'prepared'/ "pooled_candidates.csv"

# =========================
# Загрузка
# =========================
candidates = pd.read_csv(candidates_path)
retrieval_nodup = pd.read_csv(retrieval_nodup_path)
gold_subset = pd.read_csv(gold_subset_path)
```

```python
retrieval_nodup['candidate_id'].nunique()
```

```python
unique_candidate_ids = retrieval_nodup[['candidate_id']].drop_duplicates()
pooled_candidates = unique_candidate_ids.merge(candidates, on='candidate_id', how='inner')
pooled_candidates.to_csv(output_pooled_candidates_path, index=False)

print("pooled_candidates:", pooled_candidates.shape)
print(f"Файл сохранён: {output_pooled_candidates_path}")
```
